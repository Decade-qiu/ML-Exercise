```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

files = ['lstm', 'transformer', 'mymodel']
names = ['lstm', 'transformer', 'SktNet']
true_values = []
predicted_values = []
markers = ['o', 'x', 's']
colors = ['r', 'y', 'g']

for i in files:
    file = open(f'{i}/tp.txt', 'r')
    lines = file.readlines()
    file.close()
    xx, yy = [], []
    for line in lines:
        x, y = map(float, line.split())
        xx.append(x)
        yy.append(y)
    true_values = xx
    predicted_values.append(yy)

# 绘制对比曲线
plt.figure(figsize=(12, 6))
plt.plot(true_values, label="Ground Truth", linestyle='-', marker='o', markersize=3, alpha=0.6, color='b',linewidth=1.5)
# plt.plot(predicted_values_, label="Prediction", linestyle='--', marker='x', markersize=4, alpha=1, color='r',linewidth=2)
for i in range(len(files)):
    plt.plot(predicted_values[i], label=names[i], linestyle='--', marker=markers[i], markersize=4, alpha=1, color=colors[i],linewidth=2)
plt.legend()
plt.xlabel("Time Step", fontsize=14)
plt.ylabel("Rental Count", fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title("Overall Bike Rental Prediction", fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 数据
algs = ['lstm', 'transformer', 'SktNet']

I96 = [
    [0.0183, 0.0162, 0.0187],  # MSE
    [0.0861, 0.0816, 0.0881]   # MAE
]

I240 = [
    [0.0286, 0.0272, 0.0318],
    [0.1145, 0.1147, 0.1212]
]

# I96和I240误差对比柱状图
plt.figure(figsize=(8, 6))
bar_width = 0.15
index = np.arange(len(algs))

# 绘制柱状图
plt.bar(index, I96[0], bar_width, label="MSE(O=96)", color='b', alpha=0.6)
plt.bar(index + bar_width, I96[1], bar_width, label="MAE(O=96)", color='r', alpha=0.6)
plt.bar(index + 2 * bar_width, I240[0], bar_width, label="MSE(O=240)", color='y', alpha=0.6)
plt.bar(index + 3 * bar_width, I240[1], bar_width, label="MAE(O=240)", color='g', alpha=0.6)

# 设置坐标轴和标题
plt.xlabel("Algorithms", fontsize=14)
plt.ylabel("Error", fontsize=14)
plt.xticks(index + bar_width / 2, algs, fontsize=14)
plt.yticks(fontsize=14)
plt.title("I96 Error Comparison", fontsize=16)

# 调整图例位置
plt.legend(fontsize=12, bbox_to_anchor=(0.6, 1.01), loc='upper left')

# 显示图像
plt.tight_layout()
plt.show()
```
