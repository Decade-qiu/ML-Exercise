```python
# Re-import libraries and reload the dataset due to reset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# File paths for the datasets
train_file_path = 'train_data.csv'

# Load the train dataset
train_data = pd.read_csv(train_file_path)

train_data = train_data.drop(columns=['dteday'])
train_data = train_data.drop(columns=['instant'])


# Calculate the correlation matrix for the train dataset
correlation_matrix = train_data.corr()

# Plot the heatmap of correlations
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True, xticklabels=True, yticklabels=True)
plt.title("Feature Correlation Matrix")
plt.show()

# Sort correlations with the target variable 'cnt' to assess importance
target_corr = correlation_matrix['cnt'].sort_values(ascending=False)
target_corr
```
